```python
# Get dataset by id
dataset = openml.datasets.get_dataset(44224) 

# Get the data itself as a dataframe (or otherwise)
x,y,categorical_indicator2, attribute_names2 = dataset.get_data(dataset_format="dataframe")

df2 = x

print(df2.head())

print(y)

print(categorical_indicator2)

print(attribute_names2)

df2.info()

print(df2.isna().sum().sum())
```

```python
cols = ["MEMBERSHIP_TERM_YEARS", "ANNUAL_FEES"]
df_new = df2[cols].copy()

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

for ax, col in zip(axes, cols):
    ax.boxplot(df_new[col].dropna(), showfliers=False)
    ax.set_xlabel(f"{col}")
    
fig.suptitle("Boxplots of numeric variables in \"MembershipWoes\"")  
plt.tight_layout()
plt.show()
```
